# The North Carolina Housing Market


In this project we will be analyzing the North Carolina housing market over the past 20 months using housing data provided by Zillow to conduct our research as it breaks down properties via type, median rental costs, projected rental revenue, purchaser and tenant demographics from a nationwide pool of data that can be narrowed down to a specific neighborhood if needed. 

##### First will be Importing the libraries to to run the project 

```python
import pandas as pd
import panel as pn
import altair as alt
from pathlib import Path
from vega_datasets import data
import hvplot.pandas
```

Importing the necesary data 

```python
housing_data = pd.read_csv(
    Path("/Users/amtrease/Desktop/workspace/Project 1/housing_data.csv"),
)

#Reviewing first and last 5 rows of DataFrame
display(housing_data.head())
display(housing_data.tail())
```

```python
#Calling on only the NC housing Data 
nc_housing_data = housing_data.loc[housing_data["StateName"] == "NC"]

#Reviewing the Data 
nc_housing_data.head()
```

```python
# Calling on the different column names to better slice
nc_housing_data.columns

# Creating variable housing_data_info to better splice the data how we want
housing_data_info= ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName',]
```

```python
#Creating for loop to only pull housing prices from year 2020
nc_housing_data_2020= []
for column_name in nc_housing_data.columns:
    if "2020" in column_name:
        nc_housing_data_2020.append(column_name)
print(nc_housing_data_2020)
```

```python
#Creating for loop to only pull housing prices from year 2021
nc_housing_data_2021= []
for column_name in nc_housing_data.columns:
    if "2021" in column_name:
        nc_housing_data_2021.append(column_name)
print(nc_housing_data_2021)
```

```python
#Creating variable to combine years 2020 and 2021 to chart 
nc_housing_data[housing_data_info+nc_housing_data_2020+nc_housing_data_2021]
```

```python
#Using the .drop function to get rid of unneeded columns
average_price_by_city = nc_housing_data[housing_data_info+nc_housing_data_2020+nc_housing_data_2021].drop(columns=["RegionID","SizeRank","RegionType","StateName"]).groupby("RegionName").sum()

#Displaying variable
average_price_by_city.head()
```

```python
#Creating graph to show price increases over past 2 years 
average_price_by_city.T.hvplot(
    xlongitude= "Years",
    ylatitude= "Price",
    title="North Carolina Housing Prices 2020-2021",
    rot=25,
    height=500,
    width=1000
))
```
